```python
from google.colab import files
import pandas as pd

uploaded = files.upload()
file_name = next(iter(uploaded))
df = pd.read_csv(file_name)

print(df)
```

```python
import matplotlib.pyplot as plt

df.columns = ['Beverage_categories', 'Avg_sugar']

# Create a bar chart
plt.figure(figsize=(10, 8))
bars = plt.barh(df['Beverage_categories'], df['Avg_sugar'], color='lightskyblue')

# Add the data values to the bars
for bar in bars:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height() / 2,
        f'{bar.get_width():.2f}',
        va='center',
        ha='left',
        fontsize=8,
        color='black'
    )

plt.xlabel('Average Sugar')
plt.gca().invert_yaxis()
plt.title('Average Sugar Content for Each Starbucks Beverage Category')
plt.tight_layout()

# Show the bar chart
plt.show()
```
