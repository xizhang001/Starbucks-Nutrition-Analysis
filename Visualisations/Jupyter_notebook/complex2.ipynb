```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

file_name = next(iter(uploaded))
df = pd.read_csv(file_name)

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Pairing'] = df['Food'].str.strip() + ' &\n' + df['Beverages'].str.strip()

df_sorted = df.sort_values('Calorie', ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 20))
plt.barh(df_sorted['Pairing'], df_sorted['Calorie'], color='lightcoral')

# Add labels and title
plt.xlabel('Total Calories')
plt.ylabel('Food & Beverage Pairings')
plt.title('The unique combinations of food items and beverages where the total calorie count does not exceed 500')

plt.gca().invert_yaxis()

# Show the bar chart
plt.tight_layout()
plt.show()
```
