```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

file_name = next(iter(uploaded))
df = pd.read_csv(file_name)

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_sorted = df.sort_values('Calorie', ascending=False)

plt.figure(figsize=(10, 8))

cmap = plt.cm.viridis

norm = plt.Normalize(df_sorted['Calorie'].min(), df_sorted['Calorie'].max())
colors = cmap(norm(df_sorted['Calorie']))

bars = plt.barh(df_sorted['Beverage_categories'], df_sorted['Calorie'], color=colors)

for bar in bars:
    plt.text(
        bar.get_width(),
        bar.get_y() + bar.get_height() / 2,
        f'{int(bar.get_width())}',
        va='center',
        ha='left',
        fontsize=8,
        color='black'
    )

sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm)

plt.xlabel('Calories')
plt.title('Top 10 Starbucks Beverages with the Highest Calories')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```
